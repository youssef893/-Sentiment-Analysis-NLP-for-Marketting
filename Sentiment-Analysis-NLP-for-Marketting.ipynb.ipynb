```python
import pandas as pd
import spacy
from spacy.lang.en.examples import sentences 
import re
from nltk.tokenize import  word_tokenize
import nltk
from nltk.stem import WordNetLemmatizer
import tensorflow as tf
import numpy as np
import ktrain
from ktrain import text
nltk.download('punkt')
```

```python
def remove_num_punc(text):
  res = re.sub(r'[^a-zA-Z]', ' ', text)
  return res.lower()
```

```python
def lemmitization(text):
  words = ''
  for word in text:
    words += ' ' + word.lemma_
  return words
```

```python
data = pd.read_json('/kaggle/input/video-games/Video_Games_5.json', lines=True)
data.head()
```

```python
data.shape
```

```python
print(set(data['overall']))
```

```python
data = data[['reviewText','overall']]
data.head()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
data = data[:20000]
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
data.shape
```

```python
data['reviewText'] = data['reviewText'].apply(lambda text: nlp(text))
data['reviewText'] = data['reviewText'].apply(lambda text: lemmitization(text))
```

```python
data['reviewText'] = data['reviewText'].apply(lambda text: remove_num_punc(text))
# data['reviewText'] = data['reviewText'].apply(lambda text: word_tokenize(text))
```

```python
len(data['reviewText'][0])
```

```python
(data['reviewText'][0])
```

```python
data.head()
```

```python
x_train = np.array(data['reviewText'][:15000])
y_train = np.array(data['overall'][:15000].apply(lambda x: str(x))) # int or str in lambda
x_test = np.array(data['reviewText'][15000:])
y_test = np.array(data['overall'][15000:].apply(lambda x: str(x)))
```

```python
y_train
```

```python
(x_train,  y_train), (x_test, y_test), preproc = text.texts_from_array(x_train=x_train, y_train=y_train, x_test=x_test, y_test=y_test, 
                                                                       class_names=['1','2','3','4','5'],
preprocess_mode='bert',maxlen=350)
```

```python
model = text.text_classifier('bert', train_data=(x_train, y_train), preproc=preproc)
learner = ktrain.get_learner(model, train_data=(x_train, y_train), val_data=(x_test, y_test), batch_size=10)
```

```python
learner.fit_onecycle(2e-5, 5)
```

```python
predictor = ktrain.get_predictor(learner.model, preproc)
     

predictor.save('sentiment')
print('MODEL SAVED')
```
